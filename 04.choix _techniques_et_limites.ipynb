Notebook pour répertorier les choix techniques et les limites repérées

"Choix techniques" : 

- Certains de nos fichiers étaient trop lourds pour être stockés sur github (stop_times.txt et la plupart des geojson). Nous avons donc décidé de les convertir en parquet (format conseillé dans les cours de Lino Galiana), permettant non seulement le stockage mais accélérant aussi le traitement.

- Pour les poi, la plupart des premiers que nous avions identifiés étaient un peu trop "larges", nous avons donc créé des fichiers plus filtrés (sorte de v2), pour affiné les lieux ciblés. Cela nous a permis d'obtenir une meilleure corrélation avec validations que leurs équivalents sans filtre. Le plus notable étant administration2 à 0.26 contre -0.04 pour administration.

- Normalisation des poi : a eu l’effet de diminuer l’importance de logement qui est à 1300 de maximum, qui est largement au dessus des autres. 

Limites : 
- Parmi les données mises à disposition par IDFM, nous manquons parfois d'information, nous avons des données de validations, mais elle ne prennent pas en compte la fraude, les correspondances... Nous ne savons pas non plus où descendent les gens
- Concernant les poi, certains valent "plus" que d'autres, par exemple un terrain de foot loisir n'attirent pas comme le Stade de France, pourtant ils sont comptés de la même façon. 
- Problème au niveau des distances selon la zone, en plein Paris nous allons trouver beaucoup de commerce de proximité dans un petit rayon, mais dans des zones plus excentrées, "l'échelle" n'est pas la même. Il faudrait par exemple définir : hyper centre de Paris rayon 300 mètres, reste de Paris rayon 500 mètres, et banlieue 800 mètres.
- La question du sens de causalité, la présence de nombreux restaurant "prouve" la fréquentation d'un lieu mais ne l'explique pas. En revanche, la présence d'une salle de concert, d'un stade ou d'un monument justifie l'attractivité. 
- limite des noeuds : est-ce l’offre qui provoque la demande ou la demande qui provoque l’offre
